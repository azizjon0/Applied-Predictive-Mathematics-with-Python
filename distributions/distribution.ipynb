## 1. Normal Distribution — Height of Men in the UK

- **Mean (μ):** 175.3 cm  
- **Standard deviation (σ):** 7.5 cm  
- Assume a normal distribution.

### Questions:
- What percentage of men have a height **between 175.3 and 182.8 cm**?
- What is the probability that a man is **taller than 175.3 cm**?
- Is a height of **200 cm considered an outlier**?

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd

# Section 1: Normal distribution (height)
mu_height = 175.3
sigma_height = 7.5

z_1 = (182.8 - mu_height) / sigma_height  # = 1
p_between = stats.norm.cdf(z_1) - stats.norm.cdf(0)

p_greater = 1 - stats.norm.cdf(0)

z_200 = (200 - mu_height) / sigma_height
p_200 = 1 - stats.norm.cdf(z_200)
is_200_exceptional = z_200 > 3


print(f"1. P(175.3 < X < 182.8): {p_between:.4f} → {p_between * 100:.2f}%")
print(f"2. P(X > 175.3): {p_greater:.4f} → {p_greater * 100:.2f}%")
print(f"3. P(X > 200): {p_200:.5f} → {p_200 * 100:.3f}%")
print(f"   Is 200 cm exceptional? {'Yes ✅' if is_200_exceptional else 'No ❌'}\n")
```

## 2. Normal Distribution — Weight of Flour Packages

- **Labelled weight:** 1000 g  
- **Machine average (μ):** 1004 g  
- **Standard deviation (σ):** 2 g  
- Assume a normal distribution.

### Questions:
- What percentage of packages weigh **more than 1006 g**?
- What percentage weigh **less than 1000 g** (underweight)?
- Is a package of **1009 g an anomaly**?

```python
# Section 2: Normal distribution (flour)
mu_flour = 1004
sigma_flour = 2

z_more_1006 = (1006 - mu_flour) / sigma_flour
p_more_1006 = 1 - stats.norm.cdf(z_more_1006)

z_less_1000 = (1000 - mu_flour) / sigma_flour
p_less_1000 = stats.norm.cdf(z_less_1000)

z_1009 = (1009 - mu_flour) / sigma_flour
p_1009 = 1 - stats.norm.cdf(z_1009)
is_1009_exceptional = z_1009 > 2
print(f"1. P(X > 1006): {p_more_1006:.4f} → {p_more_1006 * 100:.2f}%")
print(f"2. P(X < 1000): {p_less_1000:.4f} → {p_less_1000 * 100:.2f}%")
print(f"3. P(X > 1009): {p_1009:.4f} → {p_1009 * 100:.2f}%")
print(f"   Is 1009g exceptional? {'Yes ✅' if is_1009_exceptional else 'No ❌'}\n")
```


## 3. Poisson Distribution — Illness Absences

- On average, **2 people are absent** due to illness each day.  
- Assume a **Poisson distribution**.

### Questions:
- What is the probability that **exactly 2** people are absent?
- What is the probability that **at least 3** people are absent?
- What is the probability that **fewer than 4** people are absent?

```python
# Section 3: Poisson distribution (illness)
lambda_poisson = 2
p_2 = stats.poisson.pmf(2, lambda_poisson)
p_min_3 = 1 - sum([stats.poisson.pmf(k, lambda_poisson) for k in [0, 1, 2]])
p_less_4 = stats.poisson.cdf(3, lambda_poisson)
print(f"1. P(X = 2): {p_2:.4f} → {p_2 * 100:.2f}%")
print(f"2. P(X ≥ 3): {p_min_3:.4f} → {p_min_3 * 100:.2f}%")
print(f"3. P(X < 4): {p_less_4:.4f} → {p_less_4 * 100:.2f}%\n")
```

## 4. Uniform Distribution — Continuous Variable

- Let **X ∼ U(1, 5)** — a continuous uniform distribution between 1 and 5.

### Questions:
- Calculate: **P(2 < X < 3)**
- Calculate: **P(1.5 < X < 4.5)**
- Calculate: **P(X > 2.5)**
- Calculate: **P(X ≤ 4)**

```python
# Section 4: Uniform distribution U(1, 5) 
a, b = 1, 5
density = 1 / (b - a)

p_2_3 = (3 - 2) * density
p_1_5_4_5 = (4.5 - 1.5) * density
p_greater_2_5 = (5 - 2.5) * density
p_less_equal_4 = (4 - 1) * density
print(f"1. P(2 < X < 3): {p_2_3:.2f}")
print(f"2. P(1.5 < X < 4.5): {p_1_5_4_5:.2f}")
print(f"3. P(X > 2.5): {p_greater_2_5:.3f}")
print(f"4. P(X ≤ 4): {p_less_equal_4:.2f}")
```

```python
# Plotting all four distribution visualizations

# Setup
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
x1 = np.linspace(150, 200, 500)
x2 = np.linspace(990, 1015, 500)
x3 = np.arange(0, 10)
x4 = np.linspace(1, 5, 500)

# Normal Distribution - Height
y1 = stats.norm.pdf(x1, 175.3, 7.5)
axes[0, 0].plot(x1, y1, label='PDF')
axes[0, 0].fill_between(x1, y1, where=(x1 > 175.3) & (x1 < 182.8), alpha=0.4, color='green', label='175.3–182.8 cm')
axes[0, 0].set_title('Normal Distribution (Height)')
axes[0, 0].set_xlabel('Height (cm)')
axes[0, 0].set_ylabel('Density')
axes[0, 0].legend()

# Normal Distribution - Flour
y2 = stats.norm.pdf(x2, 1004, 2)
axes[0, 1].plot(x2, y2, label='PDF')
axes[0, 1].fill_between(x2, y2, where=(x2 > 1006), alpha=0.4, color='orange', label='>1006g')
axes[0, 1].fill_between(x2, y2, where=(x2 < 1000), alpha=0.4, color='red', label='<1000g')
axes[0, 1].set_title('Normal Distribution (Flour Weight)')
axes[0, 1].set_xlabel('Weight (g)')
axes[0, 1].set_ylabel('Density')
axes[0, 1].legend()

# Poisson Distribution - Illness
y3 = stats.poisson.pmf(x3, 2)
axes[1, 0].bar(x3, y3, alpha=0.7, color='purple')
axes[1, 0].set_title('Poisson Distribution (Illness)')
axes[1, 0].set_xlabel('Number of People Sick')
axes[1, 0].set_ylabel('Probability')

# Uniform Distribution U(1,5)
y4 = [0.25] * len(x4)
axes[1, 1].plot(x4, y4, label='PDF', color='blue')
axes[1, 1].fill_between(x4, y4, where=(x4 > 2) & (x4 < 3), alpha=0.4, color='green', label='2 < X < 3')
axes[1, 1].set_ylim(0, 0.3)
axes[1, 1].set_title('Uniform Distribution U(1,5)')
axes[1, 1].set_xlabel('X')
axes[1, 1].set_ylabel('Density')
axes[1, 1].legend()

plt.tight_layout()
plt.show()
```
